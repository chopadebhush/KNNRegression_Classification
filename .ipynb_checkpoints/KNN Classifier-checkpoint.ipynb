```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.datasets import load_breast_cancer
data =load_breast_cancer()
```

```python
data
```

```python
data.target
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data=data.data ,columns=data.feature_names)
df.head()
```

```python
data.target_names
```

```python
df["target"] =data.target
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
X =df.drop("target",axis=1) ### Drop target columns
y =df["target"] ## put target columns as y var
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test ,y_train,y_test =train_test_split(X ,y,test_size=0.33,random_state=42)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.neighbors import KNeighborsClassifier
neighbours =KNeighborsClassifier()
neighbours.fit(X_train ,y_train)
y_pred =neighbours.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(confusion_matrix(y_test ,y_pred))
```

```python
print(classification_report(y_test ,y_pred))
```

```python
print(accuracy_score(y_test ,y_pred))
```

```python
# lets do hyperparameter tuning
```

```python
from sklearn.model_selection import GridSearchCV
clf =KNeighborsClassifier()
param ={"n_neighbors":np.arange(1,100),
       "weights":["uniform","distance"],
       "algorithm":["ball_tree","kd_tree"]}
```

```python
classifier =GridSearchCV(estimator=clf ,param_grid=param ,cv=5,verbose=3)
classifier.fit(X_train ,y_train)
```

```python
classifier.best_score_
```

```python
classifier.best_params_
```
